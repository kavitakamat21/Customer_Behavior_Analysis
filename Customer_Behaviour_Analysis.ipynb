#### Fixing the category :-

```python
corect_mapping = {
    # Clothing 
    "T-shirt" : "Clothing", "Shirt" : "Clothing", "Shorts" : "Clothing", "Hoodie": "Clothing",
    "Pants" : "Clothing",   "Socks" : "Clothing", "Jeans" : "Clothing", "Dress": "Clothing",
    "Blouse" : "Clothing", "Skirt" : "Clothing", "Sweater" : "Clothing", 

   # Accessories
   "Sunglasses" : "Accessories", "Gloves" : "Accessories","Jewelry" : "Accessories",
    "Hat" : "Accessories", "Handbag" : "Accessories","Backpack" : "Accessories",
    "Belt" : "Accessories", "Scarf" : "Accessories","Bag" : "Accessories",

  # Electronics
    "Laptop" : "Electronics",  "Phone" : "Electronics",  "Watch" : "Electronics", 
    "Headphones" : "Electronics",

 # Footwear
    "Shoes" : "Footwear", "Sandals" : "Footwear", "Sneakers" : "Footwear", "Boots" : "Footwear",

  # Outerwear
    "Coat" : "Outerwear", "Jacket" : "Outerwear",
   
}
```

##### Now all Electronic size is replace by "Not applicable"

```python
# Treating null values or replacing all values in Accessories
df.loc[df['Category']=='Accessories',"Size"] = "Free Size"
```

```python
# Treating null values or replacing all values in Footwear
df.loc[df['Category']=='Footwear',"Size"] = "Not Available"

#Entire columns pe apply horah tha 
```

```python
df.groupby('Category')['Size'].unique()
```

```python
# Now treating Clothing Category nan value 
# We cannot use mean and median bcoz clothing is text data --- we will replace by mode [more time repepated size]
```

```python
# Treating null values or replacing all values in Clothing
df.loc[(df['Category']=='Clothing') & (df["Size"].isnull()),"Size"] = df[df['Category']=='Clothing']['Size'].mode()[0]

#[0]  meams suppose small or medium have same number of size then pick first postions
```

```python
df.groupby('Category')['Size'].unique()
```

#### Review Rating :-

```python
df.head()
```

```python
# Now check review ratings konse value kiten hai 
df.groupby('Item Purchased')['Review Rating'].mean()
```

```python
# Sub-Category get avg ratings
```

```python
# we will replace by avg raitngs by each subcategory
# Fillings null values of Review ratings with the mean of each subcategory purcahsed

df['Review Rating'] = df['Review Rating'].fillna(df.groupby('Item Purchased')['Review Rating'].transform('mean'))
```

#### Previous Purchases :-

```python
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

#### Purchase Amount (USD):-


```python
# Treat with konse item ka kya purcahse price h usse avg se fill kr dena hai 
```

```python
df.groupby('Item Purchased')["Purchase Amount (USD)"].mean()
```

```python
df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].fillna(df.groupby('Item Purchased')["Purchase Amount (USD)"].transform('mean'))
```
